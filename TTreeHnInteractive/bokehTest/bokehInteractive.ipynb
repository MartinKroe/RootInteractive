```python
import numpy as np # we will use this later, so import it now
from bokeh.layouts import gridplot
from bokeh.models import ColumnDataSource, CDSView, BooleanFilter
from bokeh.plotting import *
from bokeh.io import output_notebook, show, push_notebook
from ipywidgets import *
from IPython.display import display
import pandas as pd
output_notebook()
```

```python
n = 1000
x = np.random.standard_normal(n)
y = np.random.standard_normal(n)
z = np.random.standard_normal(n)
data = {'x': x,'y': y,'z':z}
```

```python
df=pd.DataFrame.from_dict(data)
source = ColumnDataSource(data=data)
tools = ["box_select", "hover", "reset"]
p = figure(plot_height=500, plot_width=500, tools=tools)
plot=p.circle(x="x", y="y", size=3, hover_color="red", source=source)
handle=show(p,notebook_handle=True)

def update(cutx=0,cuty=0, cutz=0):
    global query
    query='x>'+str(cutx)+'& y>'+str(cuty) + '& z>'+str(cutz)
    print(query)
    dataNew=df.query(query)
    newSource=ColumnDataSource(data=dataNew)
    source.data=newSource.data
    push_notebook(handle)
 
interact(update,cutx=(-3,3,0.1), cuty=(-3,3,0.1),  cutz=(-3,3,0.1))  
```
