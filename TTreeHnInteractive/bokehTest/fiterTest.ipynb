```python
import numpy as np # we will use this later, so import it now
from bokeh.layouts import gridplot
from bokeh.models import ColumnDataSource, CDSView, BooleanFilter
from bokeh.plotting import figure, show
from bokeh.io import output_notebook, show
output_notebook()

n = 100000
x = np.random.standard_normal(n)
y = np.random.standard_normal(n)
data = {'x': x,
        'y': y}
source = ColumnDataSource(data=data)
#tds=TableDataSource(data=data)

#source = ColumnDataSource(data=dict(x=[1, 2, 3, 4, 5], y=[1, 2, 3, 4, 5]))


booleans = [True if y_val**2 < 1 else False for y_val  in source.data['y'] ]
#source[source['x'] > source['y']]
booleans = [True if y_val**2 < 1 else False for y_val  in source.data['y'] ]
view = CDSView(source=source, filters=[BooleanFilter(booleans)])

tools = ["box_select", "hover", "reset"]
p = figure(plot_height=300, plot_width=300, tools=tools)
p.circle(x="x", y="y", size=3, hover_color="red", source=source)
#p.circle(x="x", y="y", size=3, hover_color="red", source=source[source['x'] > source['y']])

p_filtered = figure(plot_height=300, plot_width=300, tools=tools,
                    x_range=p.x_range, y_range=p.y_range)
p_filtered.circle(x="x", y="y", size=3, hover_color="red", source=source, view=view)

show(gridplot([[p, p_filtered]]))
```
